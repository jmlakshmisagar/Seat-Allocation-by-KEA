```python
import pdfplumber
import pandas as pd
import re

pdf_path = "PGCET_cutoff_2023_mock_mcaenglish.pdf"
data = []

with pdfplumber.open(pdf_path) as pdf:
    for page_num, page in enumerate(pdf.pages, start=1):
        text = page.extract_text()
        text = re.sub(r'\s+', ' ', text)
        pattern = r"(C\d{3})\s+([A-Z][A-Za-z\s&',\-\.]+)"
        matches = re.findall(pattern, text)
        data.extend(matches)

if data:
    df = pd.DataFrame(data, columns=["College Code", "College Name"])
    csv_path = "college_codes_and_names.csv"
    df.to_csv(csv_path, index=False)
    print(f"Data extracted and saved to {csv_path}")
else:
    print("No matches found. Check if the PDF is formatted correctly or if the regex needs adjusting.")
```

```python
import pandas as pd

d2['Base College Code'] = d2['College Code'].str[:4]
d1['College Code'] = d1['College Code'].str.strip().str.upper()
d2['Base College Code'] = d2['Base College Code'].str.strip().str.upper()
filtered_d1 = d1[d1['College Code'].isin(d2['Base College Code'])]
print(filtered_d1)
```
