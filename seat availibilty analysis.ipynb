```python
import pandas as pd
```

```python
df=pd.read_csv('college_seats_cleaned.csv')
```

```python
df
```

```python
categories = ['1G', '1H', '2AG', '2AH', '2BG', '2BH', '3AG', '3AH',
              '3BG', '3BH', 'GM', 'GMH', 'NKN', 'PH', 'SCG', 'SCH', 'STG', 'STH', 'XD']

df['Sum'] = df[categories].sum(axis=1)
filtered_df = df[df['Sum'] == df['Total Seats']]
filtered_df = filtered_df.drop(columns=['Sum'])

filtered_df
```

```python
filtered_df.to_csv('final_seat_availability.csv', index=False)
```
